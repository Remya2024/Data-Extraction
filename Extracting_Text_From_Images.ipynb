```python
import os
import pandas as pd
import easyocr

# Path to the folder containing images
folder_path = '/content/drive/MyDrive/Data Source/number_plates/'

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'])

# Initialize an empty list to store extracted text
data = []

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    # Check if the file is an image (you can add other extensions if needed)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(folder_path, filename)  # Get the full path of the image

        # Extract text from the image
        result = reader.readtext(image_path)

        # Format the extracted text
        extracted_text = ' '.join([text[1] for text in result])

        # Append the filename and extracted text to the data list
        data.append({'Car Number': extracted_text})

# Create a DataFrame from the data list
df = pd.DataFrame(data)
df
```

```python
# prompt: remove space from each row of this df without using user defined functions

df['Car Number'] = df['Car Number'].str.replace(' ', '')
df
```

```python
# prompt: extract fourth and fifth element of column Car Number to a separate column

# Assuming 'df' is your DataFrame from the previous code

# Extract the fourth and fifth characters of 'Car Number'
df['State Code'] = df['Car Number'].str[3:5]
df['Unique Code'] = df['Car Number'].str[-4:]
# Display the updated DataFrame
df
```

```python
# Count the occurrences of each state code
state_code_counts = df['State Code'].value_counts()

# Display the state code counts
state_code_counts
```
